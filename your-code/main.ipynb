#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

#### How does the bin size affect your plot?

```python
"""
-The plot with 10 bins is a practic simplification, however, it leaves behinf infomation on very young babies
-The plot with 50 bins is informative, but probably some groups can be clustered for simplification.
For example, the 20 to 24 yo, where there are big differences in height that unless is our specific goal,
distract the observer.
- The plot with 20 bins is a good compromise, where the babies column appear,
we can see the drop in age after 40yo,
and there are not many distracting fluctuations between similar values.
"""
```
